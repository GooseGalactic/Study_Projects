## Предобработка данных

```python
df.info()
```

*  В первую очередь обработаем пропуски в столбцах. Пропуски в столбцах `ceiling_height`, `floors_total`, `airports_nearest`, `cityCenters_nearest`, `parks_around3000`, `parks_nearest`, `ponds_around3000`, `ponds_nearest`, `days_exposition` было решено оставить на своих местах и не удалять/заполнять значения. Такие пропуски скорее всего показывают, что данных по конкретной квартире нет, а следовательно заменять их нельзя.   
*  В столбце `locality_name` мы удаляем пропуски, так как без них наше исследование не имеет значения.  
*  В свою очередь пропуски в столбцах `is_apartment`, `balcony` можно обработать и заполнить подходящими значениями.  
*  Пропуски в `living_area` и `kitchen_area` заполним, используя медианный коэффициент, с которым площадь кухни или жилая площадь относятся к общей площади. Таким образом, можно было бы домножать этот коэффициент на общую площадь.  

```python
df.dropna(subset = ['locality_name'], inplace = True) #Удаляем квартиры с пустыми значениями местоположения
df.reset_index(drop = True, inplace = True)
df.head()
```

```python
df.locality_name.value_counts()
```

```python
df.is_apartment.value_counts()
```

Предполагаем, что пропущенные значения в столбце `is_apartment` на самом деле являются `False`, так как информацию о том, является ли квартира апартаментами обязаны указывать, но при этом это не работает в обратную сторону. Заменим NaN на False.

```python
df.is_apartment = df.is_apartment.fillna(False)
df.is_apartment.value_counts()
```

```python
df.balcony.value_counts()
```

Аналогично предполагаем, что пропущенные значения в столбце `balcony` предполагает, что балкон у квартиры отсутствует, раз его не указали. Заменим NaN на 1.

```python
df.balcony = df.balcony.fillna(0)
df.balcony.value_counts()
```

Далее разберемся с типом данных. В столбце `first_day_exposition` изменим формат даты на `datetime`. Где возможно и уместно, изменим формат данных с `float` на `int` для облегчения работы.

```python
df.first_day_exposition = pd.to_datetime(df.first_day_exposition, format = '%Y-%m-%dT%H:%M:%S') #Переводит данные из типа 
df.head()                                                                                       #object в datetime
```

```python
df[['last_price', 'balcony']] = df[['last_price', 'balcony']].astype(int) #Передаем лист из названий столбцов 
df.info()
```

```python
df[df.living_area.isna() & df.kitchen_area.isna()]
```

```python
living_area_median = (df.living_area / df.total_area).median()
df.living_area = df.living_area.fillna(value = (df.total_area * living_area_median)) #Умножаем значение total_area на медианную
df.living_area.isna().sum()                                                          #долю жилой площади
```

```python
kitchen_area_median = (df.kitchen_area / df.total_area).median()
df.kitchen_area = df.kitchen_area.fillna(value = (df.total_area * kitchen_area_median))
df.kitchen_area.isna().sum()
```

```python
df.duplicated().sum() #Проверка на дупликаты после обработки пропусков
```

### Вывод

Были обработаны пропуски в столбцах. Большая их часть была оставлена в покое, так как они указывают на отстутствие информации по данной квартире. Были заменены типы данных на `int` там, где это было осмысленно и где позволяло отсутствие нулевых значений.

## Расчёты и добавление результатов в таблицу

### Цена квадратного метра

```python
df['price_per_meter'] = df.last_price / df.total_area
df.price_per_meter = df.price_per_meter.astype('int')
df.head()
```

### День недели, месяц и год публикации объявления

```python
df['day_of_the_week'] = df.first_day_exposition.dt.dayofweek #Используя свойства DayTime, достаем из столбца нужные значения
df['publication_month'] = df.first_day_exposition.dt.month
df['publication_year'] = df.first_day_exposition.dt.year
df.head()
```

### Этаж квартиры

```python
def floor_group(row):                                 # Создаем функцию, которая присваивает то или иное значение в зависимости
    if row['floor'] == 1:                             # от значения других столбцов
        return 'первый'
    if row['floor'] == row['floors_total']:
        return 'последний'
    return 'другой'
df['floor_group'] = df.apply(floor_group, axis = 1)  # В аргументах указываем, что функция применяется по строкам
df.head()
```

### Cоотношение жилой и общей площади, отношение площади кухни к общей

```python
df['living_to_total_area'] = (df.living_area / df.total_area)
df['living_to_total_area'] = df['living_to_total_area'].round(2) #Округляем значения до второго знака после запятой
df.head()
```

```python
df['kitchen_to_total_area'] = (df.kitchen_area / df.total_area)
df['kitchen_to_total_area'] = df['kitchen_to_total_area'].round(2)
df.head()
```

### Вывод

Были добавлены столбцы, которые помогут в анализе и визуализации результатов далее. Из столбца `first_day_exposition` были извлечены день недели, месяц и год продажи квартиры. Была произведена категоризация этажей домов. Были посчитаны такие метрики, как соотношение площадей кухни, общей и жилой.
